# Resampling Technique

# We will start by separating the class that will be class 0 and class 1.

```python
# class count
class_count_0, class_count_1 =Telecommunication['churn'].value_counts()

# Separate class
class_0 = Telecommunication[Telecommunication['churn'] == 0]
class_1 = Telecommunication[Telecommunication['churn'] == 1]# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
class_0.shape
```

```python
class_1.shape
```

# 1. Random Under-Sampling

Undersampling can be defined as removing some observations of the majority class. This is done until the majority and minority class is balanced out.

```python
class_0_under = class_0.sample(class_count_1)
test_under = pd.concat([class_0_under, class_1], axis=0)

print('Random under-sampling:')
print(test_under.churn.value_counts())

test_under.churn.value_counts().plot(kind='bar', title='Count (churn)');
```

# 2.Random over-sampling

Oversampling can be defined as adding more copies to the minority class. Oversampling can be a good choice when you donâ€™t have a ton of data to work with.

```python
class_1_over = class_1.sample(class_count_0, replace=True)
test_over = pd.concat([class_0, class_1_over], axis=0)

print('Random over-sampling:')
print(test_over.churn.value_counts())

test_over.churn.value_counts().plot(kind='bar', title='Count (target)');
```

# 4. Random over-sampling with imblearn

One way to fight imbalance data is to generate new samples in the minority classes. The most naive strategy is to generate new samples by randomly sampling with replacement of the currently available samples. The RandomOverSampler offers such a scheme.

```python
# import library
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(X, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_ros))
```

# 5. Under-sampling: Tomek links

Tomek links are pairs of very close instances but of opposite classes. Removing the instances of the majority class of each pair increases the space between the two classes, facilitating the classification process.

note--Tomekâ€™s link exists if the two samples are the nearest neighbors of each other

```python
# import library
from imblearn.under_sampling import TomekLinks

tl = RandomOverSampler(sampling_strategy='majority')

# fit predictor and target variable
x_tl, y_tl = ros.fit_resample(X, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_ros))
```

# 7. NearMiss

NearMiss is an under-sampling technique. Instead of resampling the Minority class, using a distance, this will make the majority class equal to the minority class.

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss()

x_nm, y_nm = nm.fit_resample(X, y)

print('Original dataset shape:', Counter(y))
print('Resample dataset shape:', Counter(y_nm))
```

```python
step_factor = 0.05 
threshold_value = 0.2 
roc_score=0
predicted_proba = rf_model.predict_proba(X_test) #probability of prediction
while threshold_value <=0.8: #continue to check best threshold upto probability 0.8
    temp_thresh = threshold_value
    predicted = (predicted_proba [:,1] >= temp_thresh).astype('int') #change the class boundary for prediction
    print('Threshold',temp_thresh,'--',roc_auc_score(y_test, predicted))
    if roc_score<roc_auc_score(y_test, predicted): #store the threshold for best classification
        roc_score = roc_auc_score(y_test, predicted)
        thrsh_score = threshold_value
    threshold_value = threshold_value + step_factor
print('---Optimum Threshold ---',thrsh_score,'--ROC--',roc_score)
```
